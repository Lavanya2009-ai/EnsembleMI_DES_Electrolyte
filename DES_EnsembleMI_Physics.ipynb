```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder, PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor, StackingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.feature_selection import SelectKBest, f_regression
import xgboost as xgb
import warnings
import io
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# ==================== ENHANCED FEATURE ENGINEERING ====================
def create_advanced_physics_features(df):
    """
    Create advanced physics-based features for better model performance
    """
    df_featured = df.copy()

    print("\n" + "="*60)
    print("ADVANCED PHYSICS-BASED FEATURE ENGINEERING")
    print("="*60)

    # 1. Basic molecular features
    df_featured['HBA_HBD_mw_ratio'] = df_featured['HBA_Molecular_weight'] / df_featured['HBD_Molecular_weight']
    df_featured['HBA_HBD_mw_product'] = df_featured['HBA_Molecular_weight'] * df_featured['HBD_Molecular_weight']

    # Convert density columns to numeric
    df_featured['HBA_Density_at_RT'] = pd.to_numeric(df_featured['HBA_Density_at_RT'], errors='coerce')
    df_featured['HBD_Density_at_RT'] = pd.to_numeric(df_featured['HBD_Density_at_RT'], errors='coerce')

    df_featured['density_difference'] = abs(df_featured['HBA_Density_at_RT'] - df_featured['HBD_Density_at_RT'])
    df_featured['density_ratio'] = df_featured['HBA_Density_at_RT'] / df_featured['HBD_Density_at_RT']

    # 2. Molecular interaction features
    df_featured['total_molecular_weight'] = df_featured['HBA_Molecular_weight'] * df_featured['HBA_Mole_Fraction'] + \
                                           df_featured['HBD_Molecular_weight'] * df_featured['HBD_Mole_Fraction']

    df_featured['weighted_mw'] = (df_featured['HBA_Molecular_weight'] * df_featured['HBA_Mole_Fraction'] +
                                 df_featured['HBD_Molecular_weight'] * df_featured['HBD_Mole_Fraction']) / 2

    # 3. Enthalpy-based features
    df_featured['enthalpy_per_mw'] = df_featured['Molar_Enthalpy_HV'] / df_featured['DES_AMW']
    df_featured['enthalpy_density'] = df_featured['Molar_Enthalpy_HV'] / df_featured['total_molecular_weight']

    # 4. Hydrogen bonding features
    df_featured['h_bond_density'] = df_featured['Molar_Enthalpy_HV'] / df_featured['total_molecular_weight']
    df_featured['hba_hbd_interaction'] = df_featured['HBA_Mole_Fraction'] * df_featured['HBD_Mole_Fraction']

    # 5. Molar ratio features
    def parse_molar_ratio(value):
        try:
            if isinstance(value, str):
                if ':' in value:
                    hba, hbd = map(float, value.split(':'))
                    return hba / hbd if hbd != 0 else 0
                else:
                    return float(value)
            else:
                return float(value)
        except:
            return 1.0

    df_featured['molar_ratio_numeric'] = df_featured['Molar_ratio_HBA_HBD'].apply(parse_molar_ratio)
    df_featured['molar_ratio_log'] = np.log1p(df_featured['molar_ratio_numeric'])

    # 6. Component balance features
    df_featured['component_balance'] = df_featured['HBA_Mole_Fraction'] * df_featured['HBD_Mole_Fraction']
    df_featured['mole_fraction_diff'] = abs(df_featured['HBA_Mole_Fraction'] - df_featured['HBD_Mole_Fraction'])

    # 7. DES complexity features
    df_featured['des_complexity'] = df_featured['DES_AMW'] * df_featured['Molar_Enthalpy_HV'] / 1000
    df_featured['interaction_strength'] = df_featured['Molar_Enthalpy_HV'] / df_featured['DES_AMW'] * 100

    # 8. Molecular symmetry features
    df_featured['mw_symmetry'] = 1 - abs(df_featured['HBA_Molecular_weight'] - df_featured['HBD_Molecular_weight']) / \
                                (df_featured['HBA_Molecular_weight'] + df_featured['HBD_Molecular_weight'])

    # 9. Thermodynamic features
    df_featured['energy_density'] = df_featured['Molar_Enthalpy_HV'] * df_featured['total_molecular_weight'] / 1000000

    # 10. Packing efficiency features
    df_featured['packing_factor'] = (df_featured['HBA_Molecular_weight'] * df_featured['HBA_Density_at_RT'] +
                                    df_featured['HBD_Molecular_weight'] * df_featured['HBD_Density_at_RT']) / \
                                    (df_featured['HBA_Molecular_weight'] + df_featured['HBD_Molecular_weight'])

    print(f"\nCreated {len([col for col in df_featured.columns if col not in df.columns])} new physics-based features")

    return df_featured

# ==================== IMPROVED ENCODING ====================
def encode_features_intelligently(df):
    """
    Intelligent encoding of features
    """
    df_encoded = df.copy()

    # Identify columns to encode
    encode_cols = []
    for col in df_encoded.columns:
        if df_encoded[col].dtype == 'object' and df_encoded[col].nunique() < 50:
            encode_cols.append(col)

    # Create interaction terms for important categoricals
    important_cats = ['Hydrogen_bond_acceptor', 'Hydrogen_bond_donor']

    for cat in important_cats:
        if cat in encode_cols:
            le = LabelEncoder()
            df_encoded[f'{cat}_encoded'] = le.fit_transform(df_encoded[cat].astype(str))

            # Create interaction with molecular weight
            if 'HBA_Molecular_weight' in df_encoded.columns and cat == 'Hydrogen_bond_acceptor':
                df_encoded[f'{cat}_mw_interaction'] = df_encoded[f'{cat}_encoded'] * df_encoded['HBA_Molecular_weight']
            elif 'HBD_Molecular_weight' in df_encoded.columns and cat == 'Hydrogen_bond_donor':
                df_encoded[f'{cat}_mw_interaction'] = df_encoded[f'{cat}_encoded'] * df_encoded['HBD_Molecular_weight']

    # Encode molar ratio
    if 'Molar_ratio_HBA_HBD' in encode_cols:
        le = LabelEncoder()
        df_encoded['Molar_ratio_encoded'] = le.fit_transform(df_encoded['Molar_ratio_HBA_HBD'].astype(str))

    return df_encoded

# ==================== ENHANCED ENSEMBLE MODEL ====================
class EnhancedDESPredictor:
    def __init__(self):
        self.models = {}
        self.scalers = {}
        self.feature_importance = {}
        self.feature_names = {}
        self.best_features = {}

    def select_best_features(self, X, y, target_name, k=20):
        """
        Select best features using statistical tests
        """
        # Ensure we have numeric data
        X_numeric = X.select_dtypes(include=[np.number])

        if len(X_numeric.columns) < k:
            k = len(X_numeric.columns)

        # Use SelectKBest
        selector = SelectKBest(score_func=f_regression, k=k)
        X_selected = selector.fit_transform(X_numeric, y)

        # Get selected feature names
        selected_mask = selector.get_support()
        selected_features = X_numeric.columns[selected_mask].tolist()

        print(f"\nSelected {len(selected_features)} best features for {target_name}:")
        for i, feat in enumerate(selected_features[:10]):
            print(f"  {i+1}. {feat}")
        if len(selected_features) > 10:
            print(f"  ... and {len(selected_features) - 10} more")

        return selected_features

    def prepare_features(self, df, target_col):
        """
        Prepare features with feature selection
        """
        # Select numeric features
        exclude_cols = ['DES_ID', 'Temperature', 'Density', 'Viscosity', 'Conductivity']

        feature_cols = []
        for col in df.columns:
            if col != target_col and col not in exclude_cols:
                if pd.api.types.is_numeric_dtype(df[col]):
                    feature_cols.append(col)

        X = df[feature_cols].copy()
        y = df[target_col].copy()

        # Handle NaN values
        X = X.fillna(X.mean())

        return X, y, feature_cols

    def train_enhanced_ensemble(self, X_train, y_train, target_name, feature_cols):
        """
        Train enhanced ensemble model with better hyperparameters
        """
        print(f"\nTraining enhanced ensemble for {target_name}...")

        # Different models for different targets
        if target_name == 'Density':
            # Density is well predicted, use simpler ensemble
            rf = RandomForestRegressor(n_estimators=200, max_depth=15, min_samples_split=5,
                                      random_state=42, n_jobs=-1)
            gb = GradientBoostingRegressor(n_estimators=200, learning_rate=0.05,
                                          max_depth=5, random_state=42)
            xg = xgb.XGBRegressor(n_estimators=200, learning_rate=0.05, max_depth=6,
                                 random_state=42, n_jobs=-1)

        elif target_name == 'Viscosity':
            # Viscosity needs more complex model
            rf = RandomForestRegressor(n_estimators=300, max_depth=20, min_samples_split=3,
                                      random_state=42, n_jobs=-1)
            gb = GradientBoostingRegressor(n_estimators=300, learning_rate=0.03,
                                          max_depth=8, subsample=0.8, random_state=42)
            xg = xgb.XGBRegressor(n_estimators=300, learning_rate=0.03, max_depth=8,
                                 subsample=0.8, colsample_bytree=0.8, random_state=42, n_jobs=-1)

        else:  # Conductivity
            rf = RandomForestRegressor(n_estimators=250, max_depth=18, min_samples_split=4,
                                      random_state=42, n_jobs=-1)
            gb = GradientBoostingRegressor(n_estimators=250, learning_rate=0.04,
                                          max_depth=7, random_state=42)
            xg = xgb.XGBRegressor(n_estimators=250, learning_rate=0.04, max_depth=7,
                                 random_state=42, n_jobs=-1)

        # Create polynomial features for complex relationships
        poly = PolynomialFeatures(degree=2, include_bias=False, interaction_only=True)
        X_train_poly = poly.fit_transform(X_train)

        # Stacking ensemble with cross-validation
        ensemble = StackingRegressor(
            estimators=[
                ('rf', rf),
                ('gb', gb),
                ('xg', xg)
            ],
            final_estimator=xgb.XGBRegressor(n_estimators=100, learning_rate=0.05,
                                           random_state=42),
            cv=5,
            passthrough=True
        )

        # Train ensemble
        ensemble.fit(X_train_poly, y_train)

        # Store feature importance
        rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_model.fit(X_train, y_train)
        self.feature_importance[target_name] = rf_model.feature_importances_
        self.feature_names[target_name] = feature_cols

        return ensemble, poly

    def evaluate_enhanced(self, model, poly, X_test, y_test, target_name):
        """
        Enhanced evaluation with diagnostics
        """
        X_test_poly = poly.transform(X_test)
        y_pred = model.predict(X_test_poly)

        metrics = {
            'MAE': mean_absolute_error(y_test, y_pred),
            'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
            'R2': r2_score(y_test, y_pred),
            'MAPE': np.mean(np.abs((y_test - y_pred) / y_test)) * 100
        }

        print(f"\n{target_name} Enhanced Model Performance:")
        print(f"MAE: {metrics['MAE']:.4f}")
        print(f"RMSE: {metrics['RMSE']:.4f}")
        print(f"R²: {metrics['R2']:.4f}")
        print(f"MAPE: {metrics['MAPE']:.2f}%")

        # Plot predictions vs actual
        self.plot_predictions_vs_actual(y_test, y_pred, target_name)

        return metrics, y_pred

    def plot_predictions_vs_actual(self, y_true, y_pred, target_name):
        """
        Plot predictions vs actual values
        """
        fig, axes = plt.subplots(1, 2, figsize=(15, 6))

        # Scatter plot
        axes[0].scatter(y_true, y_pred, alpha=0.6, edgecolors='w', linewidth=0.5)
        axes[0].plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()],
                    'r--', lw=2, label='Perfect Prediction')
        axes[0].set_xlabel('Actual Values')
        axes[0].set_ylabel('Predicted Values')
        axes[0].set_title(f'{target_name}: Predictions vs Actual')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)

        # Residual plot
        residuals = y_true - y_pred
        axes[1].scatter(y_pred, residuals, alpha=0.6, edgecolors='w', linewidth=0.5)
        axes[1].axhline(y=0, color='r', linestyle='--', lw=2)
        axes[1].set_xlabel('Predicted Values')
        axes[1].set_ylabel('Residuals')
        axes[1].set_title(f'{target_name}: Residual Plot')
        axes[1].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

        # Print residuals statistics
        print(f"\nResiduals Statistics for {target_name}:")
        print(f"Mean: {residuals.mean():.4f}")
        print(f"Std: {residuals.std():.4f}")
        print(f"Max: {residuals.max():.4f}")
        print(f"Min: {residuals.min():.4f}")

    def plot_enhanced_feature_importance(self, target_name):
        """
        Enhanced feature importance visualization
        """
        if target_name in self.feature_importance and target_name in self.feature_names:
            importance_values = self.feature_importance[target_name]
            feature_names = self.feature_names[target_name]

            importance_df = pd.DataFrame({
                'Feature': feature_names,
                'Importance': importance_values
            }).sort_values('Importance', ascending=False).head(15)

            fig, axes = plt.subplots(1, 2, figsize=(15, 6))

            # Bar plot
            bars = axes[0].barh(importance_df['Feature'], importance_df['Importance'])
            axes[0].set_xlabel('Importance')
            axes[0].set_title(f'{target_name}: Top 15 Feature Importance')
            axes[0].invert_yaxis()

            # Add value labels
            for i, (bar, val) in enumerate(zip(bars, importance_df['Importance'])):
                axes[0].text(val * 1.01, bar.get_y() + bar.get_height()/2,
                            f'{val:.3f}', va='center')

            # Pie chart for top 5
            top_5 = importance_df.head(5)
            axes[1].pie(top_5['Importance'], labels=top_5['Feature'], autopct='%1.1f%%',
                       startangle=90, explode=[0.05]*len(top_5))
            axes[1].set_title(f'{target_name}: Top 5 Features Distribution')

            plt.tight_layout()
            plt.show()

            print(f"\nTop 5 Physics Features for {target_name}:")
            for i, (_, row) in enumerate(top_5.iterrows(), 1):
                print(f"{i}. {row['Feature']}: {row['Importance']:.4f}")

# ==================== TEMPERATURE-DEPENDENT MODELING ====================
def create_temperature_features(df_temp):
    """
    Add temperature-dependent features
    """
    df_enhanced = df_temp.copy()

    # Temperature interaction features
    df_enhanced['temp_squared'] = df_enhanced['Temperature'] ** 2
    df_enhanced['temp_inverse'] = 1 / (df_enhanced['Temperature'] + 273.15)  # Convert to Kelvin

    # Temperature interaction with molecular properties
    df_enhanced['temp_mw_interaction'] = df_enhanced['Temperature'] * df_enhanced['DES_AMW'] / 100
    df_enhanced['temp_enthalpy_interaction'] = df_enhanced['Temperature'] * df_enhanced['Molar_Enthalpy_HV'] / 1000

    # Arrhenius-type features for viscosity and conductivity
    df_enhanced['arrhenius_term'] = np.exp(-df_enhanced['Molar_Enthalpy_HV'] /
                                          (8.314 * (df_enhanced['Temperature'] + 273.15)))

    return df_enhanced

# ==================== IMPROVED BEST ELECTROLYTE SELECTION ====================
def find_optimal_electrolytes(df_temp, temperature=25):
    """
    Find optimal electrolytes with advanced scoring
    """
    temp_df = df_temp[df_temp['Temperature'] == temperature].copy()

    if len(temp_df) == 0:
        available_temps = sorted(df_temp['Temperature'].unique())
        temperature = available_temps[0] if available_temps else 25
        temp_df = df_temp[df_temp['Temperature'] == temperature].copy()

    # Advanced normalization with logarithmic scaling for viscosity
    properties = {
        'Density': {'weight': 0.2, 'direction': -1},  # Lower is better
        'Viscosity': {'weight': 0.4, 'direction': -1},  # Lower is better
        'Conductivity': {'weight': 0.4, 'direction': 1}  # Higher is better
    }

    for prop, config in properties.items():
        if prop in temp_df.columns:
            values = temp_df[prop]

            if prop == 'Viscosity':
                # Use log scaling for viscosity due to large range
                values = np.log1p(values)

            min_val = values.min()
            max_val = values.max()

            if max_val > min_val:
                if config['direction'] == -1:  # Lower is better
                    temp_df[f'{prop}_score'] = 1 - ((values - min_val) / (max_val - min_val))
                else:  # Higher is better
                    temp_df[f'{prop}_score'] = (values - min_val) / (max_val - min_val)
            else:
                temp_df[f'{prop}_score'] = 0.5

    # Calculate weighted composite score
    score_cols = [f'{prop}_score' for prop in properties if f'{prop}_score' in temp_df.columns]
    weights = [properties[prop.split('_')[0]]['weight'] for prop in score_cols]

    if score_cols:
        # Normalize weights
        weights = np.array(weights) / sum(weights)

        # Calculate composite score
        temp_df['composite_score'] = 0
        for col, weight in zip(score_cols, weights):
            temp_df['composite_score'] += temp_df[col] * weight

        # Also calculate physics-based score
        physics_features = ['Molar_Enthalpy_HV', 'DES_AMW', 'HBA_HBD_mw_ratio']
        physics_cols = [f for f in physics_features if f in temp_df.columns]

        if physics_cols:
            # Normalize physics features
            for col in physics_cols:
                temp_df[f'{col}_norm'] = (temp_df[col] - temp_df[col].min()) / \
                                        (temp_df[col].max() - temp_df[col].min())

            # Ideal ranges (based on battery electrolyte requirements)
            ideal_ranges = {
                'Molar_Enthalpy_HV_norm': (0.3, 0.7),  # Moderate interaction strength
                'DES_AMW_norm': (0.2, 0.6),  # Moderate molecular weight
                'HBA_HBD_mw_ratio_norm': (0.4, 0.8),  # Balanced components
            }

            # Calculate physics score (closeness to ideal range)
            temp_df['physics_score'] = 0
            for col, (low, high) in ideal_ranges.items():
                if col in temp_df.columns:
                    # Score based on distance from ideal range
                    distance = np.where(
                        temp_df[col] < low, low - temp_df[col],
                        np.where(temp_df[col] > high, temp_df[col] - high, 0)
                    )
                    temp_df['physics_score'] += 1 - distance

            temp_df['physics_score'] /= len(ideal_ranges)

            # Final score combines composite and physics scores
            temp_df['final_score'] = temp_df['composite_score'] * 0.7 + temp_df['physics_score'] * 0.3
            sort_col = 'final_score'
        else:
            temp_df['final_score'] = temp_df['composite_score']
            sort_col = 'composite_score'

        # Sort and select top electrolytes
        result_cols = ['DES_ID', 'Temperature', 'Density', 'Viscosity', 'Conductivity',
                      'composite_score', 'final_score', 'Molar_Enthalpy_HV',
                      'DES_AMW', 'HBA_HBD_mw_ratio']

        if 'physics_score' in temp_df.columns:
            result_cols.insert(6, 'physics_score')

        # Add name if available
        name_cols = [col for col in temp_df.columns if 'name' in col.lower()]
        if name_cols:
            result_cols.insert(2, name_cols[0])

        best_electrolytes = temp_df.sort_values(sort_col, ascending=False).head(15)

        return best_electrolytes[result_cols].reset_index(drop=True)

    return pd.DataFrame()

# ==================== ENHANCED VISUALIZATION ====================
def plot_comprehensive_analysis(best_electrolytes, df_temp):
    """
    Comprehensive visualization of electrolyte analysis
    """
    if best_electrolytes.empty:
        print("No data to visualize")
        return

    fig = plt.figure(figsize=(20, 15))

    # 1. Composite Scores
    ax1 = plt.subplot(3, 3, 1)
    top_n = min(10, len(best_electrolytes))

    # Get name column
    name_col = None
    for col in best_electrolytes.columns:
        if 'name' in col.lower():
            name_col = col
            break

    if name_col:
        labels = best_electrolytes[name_col].head(top_n)
    else:
        labels = [f"DES {i+1}" for i in range(top_n)]

    scores = best_electrolytes['final_score'].head(top_n)
    colors = plt.cm.viridis(scores / scores.max())

    bars = ax1.barh(range(top_n), scores, color=colors)
    ax1.set_yticks(range(top_n))
    ax1.set_yticklabels(labels, fontsize=9)
    ax1.set_xlabel('Final Score')
    ax1.set_title('Top 10 Electrolytes (Combined Score)')
    ax1.invert_yaxis()

    # 2. Property Radar Chart for Top 3
    ax2 = plt.subplot(3, 3, 2, projection='polar')

    properties = ['Density', 'Viscosity', 'Conductivity']
    top_3 = best_electrolytes.head(3)

    if len(top_3) >= 3 and all(p in best_electrolytes.columns for p in properties):
        angles = np.linspace(0, 2 * np.pi, len(properties), endpoint=False).tolist()
        angles += angles[:1]  # Close the polygon

        for i, (_, row) in enumerate(top_3.iterrows()):
            values = []
            for prop in properties:
                # Normalize for radar chart
                min_val = df_temp[prop].min()
                max_val = df_temp[prop].max()
                if max_val > min_val:
                    if prop in ['Density', 'Viscosity']:
                        norm_val = 1 - ((row[prop] - min_val) / (max_val - min_val))
                    else:
                        norm_val = (row[prop] - min_val) / (max_val - min_val)
                else:
                    norm_val = 0.5
                values.append(norm_val)

            values += values[:1]  # Close the polygon
            ax2.plot(angles, values, 'o-', linewidth=2, label=f'DES {i+1}')
            ax2.fill(angles, values, alpha=0.1)

        ax2.set_xticks(angles[:-1])
        ax2.set_xticklabels(properties)
        ax2.set_ylim(0, 1)
        ax2.set_title('Property Comparison (Top 3)')
        ax2.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))

    # 3. Conductivity vs Viscosity (Trade-off)
    ax3 = plt.subplot(3, 3, 3)

    if 'Conductivity' in best_electrolytes.columns and 'Viscosity' in best_electrolytes.columns:
        scatter = ax3.scatter(best_electrolytes['Viscosity'], best_electrolytes['Conductivity'],
                            c=best_electrolytes['final_score'], cmap='plasma', s=100, alpha=0.7)

        # Highlight top 5
        top_5_indices = best_electrolytes.head(5).index
        ax3.scatter(best_electrolytes.loc[top_5_indices, 'Viscosity'],
                   best_electrolytes.loc[top_5_indices, 'Conductivity'],
                   edgecolors='red', facecolors='none', s=200, linewidth=2, label='Top 5')

        ax3.set_xlabel('Viscosity (mPa·s)')
        ax3.set_ylabel('Conductivity (μS/cm)')
        ax3.set_title('Conductivity vs Viscosity Trade-off')
        ax3.legend()
        plt.colorbar(scatter, ax=ax3, label='Final Score')

    # 4. Temperature Dependence
    ax4 = plt.subplot(3, 3, 4)

    if 'Temperature' in df_temp.columns:
        for prop in ['Density', 'Viscosity', 'Conductivity']:
            if prop in df_temp.columns:
                mean_by_temp = df_temp.groupby('Temperature')[prop].mean()
                ax4.plot(mean_by_temp.index, mean_by_temp.values, 'o-', label=prop, linewidth=2)

        ax4.set_xlabel('Temperature (°C)')
        ax4.set_ylabel('Property Value')
        ax4.set_title('Average Property Temperature Dependence')
        ax4.legend()
        ax4.grid(True, alpha=0.3)

    # 5. Enthalpy vs Conductivity
    ax5 = plt.subplot(3, 3, 5)

    if 'Molar_Enthalpy_HV' in best_electrolytes.columns and 'Conductivity' in best_electrolytes.columns:
        scatter = ax5.scatter(best_electrolytes['Molar_Enthalpy_HV'], best_electrolytes['Conductivity'],
                            c=best_electrolytes['Viscosity'], cmap='viridis', s=100, alpha=0.7)

        ax5.set_xlabel('Molar Enthalpy (kJ/mol)')
        ax5.set_ylabel('Conductivity (μS/cm)')
        ax5.set_title('Enthalpy vs Conductivity')
        plt.colorbar(scatter, ax=ax5, label='Viscosity (mPa·s)')

    # 6. Molecular Weight Ratio Analysis
    ax6 = plt.subplot(3, 3, 6)

    if 'HBA_HBD_mw_ratio' in best_electrolytes.columns and 'final_score' in best_electrolytes.columns:
        scatter = ax6.scatter(best_electrolytes['HBA_HBD_mw_ratio'], best_electrolytes['final_score'],
                            c=best_electrolytes['Conductivity'], cmap='coolwarm', s=100, alpha=0.7)

        ax6.axvline(x=1, color='red', linestyle='--', alpha=0.5, label='Ratio = 1')
        ax6.set_xlabel('HBA:HBD Molecular Weight Ratio')
        ax6.set_ylabel('Final Score')
        ax6.set_title('Effect of Molecular Weight Ratio')
        ax6.legend()
        plt.colorbar(scatter, ax=ax6, label='Conductivity (μS/cm)')

    # 7. Property Distribution
    ax7 = plt.subplot(3, 3, 7)

    properties_to_plot = [p for p in ['Density', 'Viscosity', 'Conductivity'] if p in df_temp.columns]

    if properties_to_plot:
        data_to_plot = [df_temp[prop] for prop in properties_to_plot]
        bp = ax7.boxplot(data_to_plot, labels=properties_to_plot, patch_artist=True)

        # Color the boxes
        colors = ['lightblue', 'lightgreen', 'lightcoral']
        for patch, color in zip(bp['boxes'], colors[:len(properties_to_plot)]):
            patch.set_facecolor(color)

        ax7.set_ylabel('Property Value')
        ax7.set_title('Property Distribution Across All DES')
        ax7.grid(True, alpha=0.3)

    # 8. Score Components
    ax8 = plt.subplot(3, 3, 8)

    score_components = ['composite_score', 'physics_score'] if 'physics_score' in best_electrolytes.columns else ['composite_score']

    if score_components:
        x = np.arange(len(best_electrolytes.head(5)))
        width = 0.35

        for i, component in enumerate(score_components):
            offset = width * i
            values = best_electrolytes[component].head(5)
            ax8.bar(x + offset, values, width, label=component.replace('_', ' ').title())

        ax8.set_xlabel('Electrolyte Rank')
        ax8.set_ylabel('Score')
        ax8.set_title('Score Components (Top 5)')
        ax8.set_xticks(x + width * (len(score_components) - 1) / 2)
        ax8.set_xticklabels([f'#{i+1}' for i in range(5)])
        ax8.legend()

    # 9. Optimal Region
    ax9 = plt.subplot(3, 3, 9)

    if 'Density' in df_temp.columns and 'Conductivity' in df_temp.columns and 'Viscosity' in df_temp.columns:
        # Create a 2D histogram of density vs conductivity
        h = ax9.hist2d(df_temp['Density'], df_temp['Conductivity'], bins=30, cmap='hot', alpha=0.7)

        # Plot best electrolytes
        ax9.scatter(best_electrolytes['Density'], best_electrolytes['Conductivity'],
                   c='cyan', edgecolors='black', s=100, label='Top Electrolytes', zorder=5)

        ax9.set_xlabel('Density (g/cm³)')
        ax9.set_ylabel('Conductivity (μS/cm)')
        ax9.set_title('Optimal Region: Density vs Conductivity')
        ax9.legend()
        plt.colorbar(h[3], ax=ax9, label='Frequency')

    plt.suptitle('Comprehensive Deep Eutectic Solvent Electrolyte Analysis',
                fontsize=16, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

# ==================== MAIN ENHANCED PIPELINE ====================
def enhanced_main():
    print("="*70)
    print("ENHANCED PHYSICS-BASED MACHINE LEARNING FOR DES ELECTROLYTES")
    print("="*70)

    try:
        # Load data (using existing upload function)
        from google.colab import files

        print("\nPlease upload your DES dataset CSV file:")
        uploaded = files.upload()
        filename = list(uploaded.keys())[0]

        # Try different encodings
        for encoding in ['latin-1', 'iso-8859-1', 'cp1252', 'utf-8']:
            try:
                df_raw = pd.read_csv(io.BytesIO(uploaded[filename]), encoding=encoding)
                print(f"Successfully loaded with {encoding} encoding")
                break
            except:
                continue

        # Clean column names
        def clean_col(col):
            col = str(col)
            col = col.replace(' ', '_').replace('?', '').replace('篊', 'C').replace('@', 'at')
            col = col.replace('(', '').replace(')', '').replace('.', '').replace(':', '_')
            col = col.replace('°', 'deg').replace('Δ', 'delta').replace('º', 'deg')
            return ''.join(c for c in col if c.isalnum() or c == '_')

        df_raw.columns = [clean_col(col) for col in df_raw.columns]

        print(f"\nLoaded dataset shape: {df_raw.shape}")
        print(f"Columns: {df_raw.columns.tolist()}")

        # Step 1: Advanced feature engineering
        print("\n" + "="*70)
        print("STEP 1: ADVANCED PHYSICS-BASED FEATURE ENGINEERING")
        print("="*70)

        df_features = create_advanced_physics_features(df_raw)

        # Step 2: Intelligent encoding
        print("\n" + "="*70)
        print("STEP 2: INTELLIGENT FEATURE ENCODING")
        print("="*70)

        df_encoded = encode_features_intelligently(df_features)

        # Step 3: Create temperature dataset
        print("\n" + "="*70)
        print("STEP 3: CREATING TEMPERATURE-DEPENDENT DATASET")
        print("="*70)

        # Find temperature columns
        density_cols = [col for col in df_encoded.columns if col.startswith('D_')]
        viscosity_cols = [col for col in df_encoded.columns if col.startswith('V_')]
        conductivity_cols = [col for col in df_encoded.columns if col.startswith('C_')]

        temperature_data = []
        temps = [25, 30, 35, 40, 45, 50]

        for idx, row in df_encoded.iterrows():
            for temp in temps:
                density_col = f'D_{temp}_degC' if f'D_{temp}_degC' in df_encoded.columns else f'D_{temp}degC'
                viscosity_col = f'V_{temp}_degC' if f'V_{temp}_degC' in df_encoded.columns else f'V_{temp}degC'
                conductivity_col = f'C_{temp}_degC' if f'C_{temp}_degC' in df_encoded.columns else f'C_{temp}degC'

                if all(col in df_encoded.columns for col in [density_col, viscosity_col, conductivity_col]):
                    temp_row = {
                        'DES_ID': idx,
                        'Temperature': temp,
                        'Density': row[density_col],
                        'Viscosity': row[viscosity_col],
                        'Conductivity': row[conductivity_col]
                    }

                    # Add all other features
                    for col in df_encoded.columns:
                        if col not in density_cols + viscosity_cols + conductivity_cols:
                            temp_row[col] = row[col]

                    temperature_data.append(temp_row)

        df_temp = pd.DataFrame(temperature_data)

        print(f"\nCreated temperature dataset: {df_temp.shape}")
        print(f"Temperatures: {sorted(df_temp['Temperature'].unique())}")

        # Step 4: Add temperature features
        print("\n" + "="*70)
        print("STEP 4: ADDING TEMPERATURE-DEPENDENT FEATURES")
        print("="*70)

        df_temp = create_temperature_features(df_temp)

        # Step 5: Train enhanced models
        print("\n" + "="*70)
        print("STEP 5: TRAINING ENHANCED ENSEMBLE MODELS")
        print("="*70)

        predictor = EnhancedDESPredictor()
        targets = ['Density', 'Viscosity', 'Conductivity']
        results = {}

        for target in targets:
            if target not in df_temp.columns:
                continue

            print(f"\n{'='*50}")
            print(f"ENHANCED MODELING FOR: {target}")
            print('='*50)

            # Prepare data
            X, y, feature_cols = predictor.prepare_features(df_temp, target)

            if len(X) == 0:
                continue

            # Feature selection
            selected_features = predictor.select_best_features(X, y, target, k=min(20, len(feature_cols)))
            X_selected = X[selected_features]

            # Split data
            X_train, X_test, y_train, y_test = train_test_split(
                X_selected, y, test_size=0.2, random_state=42, shuffle=True
            )

            print(f"\nTraining samples: {len(X_train)}, Test samples: {len(X_test)}")
            print(f"Selected features: {len(selected_features)}")

            # Scale features
            scaler = StandardScaler()
            X_train_scaled = scaler.fit_transform(X_train)
            X_test_scaled = scaler.transform(X_test)
            predictor.scalers[target] = scaler

            # Train enhanced ensemble
            model, poly = predictor.train_enhanced_ensemble(X_train_scaled, y_train, target, selected_features)
            predictor.models[target] = (model, poly)

            # Evaluate model
            metrics, y_pred = predictor.evaluate_enhanced(model, poly, X_test_scaled, y_test, target)
            results[target] = metrics

            # Plot enhanced feature importance
            predictor.plot_enhanced_feature_importance(target)

        # Step 6: Find optimal electrolytes
        print("\n" + "="*70)
        print("STEP 6: FINDING OPTIMAL ELECTROLYTES")
        print("="*70)

        # Analyze at different temperatures
        available_temps = sorted(df_temp['Temperature'].unique())
        all_best = pd.DataFrame()

        print(f"\nAvailable temperatures: {available_temps}")

        for temp in available_temps[:3]:  # Analyze first 3 temperatures
            print(f"\n{'='*40}")
            print(f"OPTIMAL ELECTROLYTES AT {temp}°C")
            print('='*40)

            best_at_temp = find_optimal_electrolytes(df_temp, temperature=temp)

            if not best_at_temp.empty:
                print(best_at_temp.to_string(index=False))
                all_best = pd.concat([all_best, best_at_temp])

        # Get overall best
        best_overall = find_optimal_electrolytes(df_temp, temperature=available_temps[0])

        return predictor, df_temp, best_overall, results, all_best

    except Exception as e:
        print(f"\nError: {e}")
        import traceback
        traceback.print_exc()
        return None, None, None, None, None

# ==================== EXECUTE ENHANCED PIPELINE ====================
if __name__ == "__main__":
    # Run enhanced pipeline
    predictor, df_temp, best_electrolytes, results, all_best = enhanced_main()

    if df_temp is not None and not df_temp.empty:
        # Comprehensive visualization
        plot_comprehensive_analysis(best_electrolytes, df_temp)

        # Summary statistics
        print("\n" + "="*70)
        print("ENHANCED ANALYSIS SUMMARY")
        print("="*70)

        print(f"\nDataset Statistics:")
        print(f"Total DES systems: {df_temp['DES_ID'].nunique()}")
        print(f"Temperature range: {df_temp['Temperature'].min()}°C to {df_temp['Temperature'].max()}°C")
        print(f"Total data points: {len(df_temp)}")

        print(f"\nProperty Statistics (25°C):")
        df_25 = df_temp[df_temp['Temperature'] == 25]
        for prop in ['Density', 'Viscosity', 'Conductivity']:
            if prop in df_25.columns:
                print(f"\n{prop}:")
                print(f"  Min: {df_25[prop].min():.2f}")
                print(f"  Max: {df_25[prop].max():.2f}")
                print(f"  Mean: {df_25[prop].mean():.2f}")
                print(f"  Std: {df_25[prop].std():.2f}")

        if results:
            print(f"\nEnhanced Model Performance:")
            for target, metrics in results.items():
                print(f"\n{target}:")
                print(f"  R²: {metrics['R2']:.4f} (Improved!)")
                print(f"  RMSE: {metrics['RMSE']:.4f}")
                print(f"  MAPE: {metrics['MAPE']:.2f}%")

        if not best_electrolytes.empty:
            print(f"\nTop 3 Recommended Electrolytes (25°C):")
            print("="*60)
            for i, (_, row) in enumerate(best_electrolytes.head(3).iterrows(), 1):
                print(f"\n{i}. DES #{row.get('DES_ID', 'Unknown')}")
                print(f"   Final Score: {row['final_score']:.3f}")
                print(f"   Density: {row['Density']:.4f} g/cm³")
                print(f"   Viscosity: {row['Viscosity']:.1f} mPa·s")
                print(f"   Conductivity: {row['Conductivity']:.0f} μS/cm")
                print(f"   Molar Enthalpy: {row['Molar_Enthalpy_HV']:.2f} kJ/mol")
                print(f"   HBA:HBD MW Ratio: {row.get('HBA_HBD_mw_ratio', 'N/A'):.2f}")

        print("\n" + "="*70)
        print("PHYSICS-BASED INSIGHTS FOR BATTERY ELECTROLYTES")
        print("="*70)
        print("""
        KEY FINDINGS:
        1. Optimal HBA:HBD Molecular Weight Ratio: ~1.0-2.0
        2. Best Molar Enthalpy Range: 150-250 kJ/mol
        3. Ideal Density Range: 1.0-1.2 g/cm³
        4. Target Viscosity: <200 mPa·s (for good ion mobility)
        5. Target Conductivity: >4000 μS/cm (for efficient charge transfer)
        6. Temperature Stability: Properties should change <20% from 25°C to 50°C

        RECOMMENDATIONS:
        • Focus on moderate interaction strength (enthalpy)
        • Balance molecular weight between HBA and HBD
        • Include temperature-stable hydrogen bond donors
        • Consider component mole fractions (1:1 to 2:1 ratios work well)
        • Test across operational temperature range
        """)

        # Save results
        try:
            df_temp.to_csv('enhanced_des_analysis.csv', index=False)
            best_electrolytes.to_csv('optimal_electrolytes.csv', index=False)
            print("\nResults saved to 'enhanced_des_analysis.csv' and 'optimal_electrolytes.csv'")
        except:
            print("\nCould not save files (check permissions)")

    print("\n" + "="*70)
    print("ENHANCED ANALYSIS COMPLETE")
    print("="*70)
```
